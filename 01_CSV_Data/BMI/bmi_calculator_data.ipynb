### 2- Unit Conversion

```python
heights_imperial = df['Height(Inches)']
# heights_imperial = df.iloc[:, 0]
weights_imperial = df['Weight(Pounds)']
# heights_imperial = df.iloc[:, 1]
```

```python
type(heights_imperial)
```

```python
heights_metric = heights_imperial * 2.54
weights_metric = weights_imperial * 0.453592
```

```python
weights_metric.describe()
```

### 3- Calculate BMI

```python
bmis = weights_metric / (heights_metric/100) ** 2
```

```python
bmis.describe()
```

### 4- Create the Result

```python
# 1- Create a dictionary
result_dict = {'Height': heights_metric.round(1),
               'Weight': weights_metric.round(1),
               'BMI': bmis.round(2)}

# 2- Convert dict to DataFrame
result_df = pd.DataFrame(result_dict)
```

```python
result_df.sort_values(by='BMI', ascending=False,
                      inplace=True, ignore_index=True)
result_df.index = df.index
```

```python
result_df.head()
```

### 5- Save the Result

```python
os.makedirs('results', exist_ok=True)
```

```python
result_df.to_csv('results/result.csv')
result_df.to_html('results/result.html')
result_df.to_excel('results/result.xlsx')
result_df.to_json('results/result.json', indent=2)
```
