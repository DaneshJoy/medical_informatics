### Create a signal

```python
x = np.linspace(0, 4*np.pi, 100)
```

```python
x.shape
```

```python
y = np.sin(x)
y2 = np.abs(np.sin(2*x))
```

```python
y.shape
```

### Visualize the signal

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 3))
plt.plot(x, y, 'm--', label='Sin(x)')
plt.plot(x, y2, 'c-.', label='|sin(2x)|')
plt.title('Signals')
plt.xlabel('t')
plt.ylabel('amplitude')
plt.legend()
```
