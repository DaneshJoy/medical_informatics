### 2- Unit Conversion

```python
heights_imperial = df['Height(Inches)']
# heights_imperial = df.iloc[:, 0]
weights_imperial = df['Weight(Pounds)']
# heights_imperial = df.iloc[:, 1]
```

```python
type(heights_imperial)
```

```python
heights_metric = heights_imperial * 2.54
weights_metric = weights_imperial * 0.453592
```

```python
weights_metric.describe()
```

### 3- Calculate BMI

```python
bmis = weights_metric / (heights_metric/100) ** 2
```
